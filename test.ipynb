```python
from ultralytics import YOLO

model = YOLO()  # or YOLO("yolo11n.yaml") / YOLO("your_model.pt")

results = model.train(
    data="coco8.yaml",
    batch= 4,
    epochs=2,
    imgsz=640,
    # device= 0,
    pretrained=False,

    # dual teacher-student (full net = teacher, subnet = student)
    dual_channel_ratio=3,
    dual_channel_loss_weight=1.0,

    teacher_distill_update= True, # (bool) if True teacher receives distillation gradients, else teacher is detached in all distillation losses

    # 1) KL classification distillation
    cls_dist=True,
    cls_alpha=1.0,
    cls_dist_t=2.0,

    # 2) M2D2 distillation
    m2d2_dist=True,
    m2d2_alpha=1.0,
    m2d2_t=2.0,

    # 3) L2 bbox distillation
    l2_dist=True,
    l2_alpha=1.0,

    # adaptive alpha controls
    adaptive_distill_alpha=True,
    adaptive_alpha_momentum=0.9,
    adaptive_alpha_min_scale=0.25,
    adaptive_alpha_max_scale=4.0,

    # masking/weighting knobs (only applied for enabled methods)
    cls_fg_mask=True,      # only matters if cls_dist=True
    # dfl_fg_mask=True,      # M2D2 now always gets TAL masks when enabled
    l2_fg_mask=True,       # used since l2_dist=True
    mask_type="pyramid",  # "original" or "pyramid"
    level_weights=[1.0, 1.0, 1.0],

    print_bn_params_dims=False,
    bn_capacity_reg= True, # (bool) enable BN-gamma suppression for teacher-only channels in dual-channel training
    bn_capacity_reg_mode= "ratio", # (str) regularization mode for teacher-only channels, choices=[ratio, l1]
    bn_capacity_reg_lambda= 1e-4, # (float) max regularization weight for BN-gamma suppression
    bn_capacity_reg_schedule= "cosine", # (str) annealing schedule for BN-gamma suppression, choices=[linear, cosine]
)
```
