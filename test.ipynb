```python
from ultralytics import YOLO

model = YOLO()  # or your model yaml/pt

results = model.train(
    data="coco8.yaml",
    epochs=2,
    imgsz=640,
    pretrained=False,

    # dual teacher-student (full net = teacher, subnet = student)
    dual_channel_ratio=3,
    dual_channel_loss_weight=1.0,

    # 1) KL classification distillation
    cls_dist=False,    # explicit on/off
    cls_alpha=1.0,     # base alpha (used only if cls_dist=True)
    cls_dist_t=2.0,

    # 2) M2D2 distillation
    m2d2_dist=False,    # explicit on/off
    m2d2_alpha=1.0,    # base alpha
    m2d2_t=2.0,

    # 3) L2 bbox distillation
    l2_dist=True,      # explicit on/off
    l2_alpha=1.0,      # base alpha

    # adaptive alpha controls (optional)
    adaptive_distill_alpha=True,
    adaptive_alpha_momentum=0.9,
    adaptive_alpha_min_scale=0.25,
    adaptive_alpha_max_scale=4.0,

    # optional masking/weighting knobs
    cls_fg_mask=True,
    dfl_fg_mask=True,
    l2_fg_mask=True,
    mask_type="original",          # "original" or "pyramid"
    level_weights=[1.0, 1.0, 1.0],

    print_bn_params_dims=True,
)
```
